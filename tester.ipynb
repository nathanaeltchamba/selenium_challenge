```python
class Player:
    _list = []
    
    def __init__(self, name, team, position, age, games_played, minutes_per_game, ft_attempts, ft_percentage, tp_attempts, tp_percentage, thp_attempts, thp_percentage, points_per_game, rebounds_per_game, assists_per_game, steals_per_game, blocks_per_game, turnovers_per_game):
        self.name = name
        self.team = team
        self.position = position
        self.age = age
        self.games_played = games_played
        self.minutes_per_game = minutes_per_game
        self.ft_attempts = ft_attempts
        self.ft_percentage = ft_percentage
        self.tp_attempts = tp_attempts
        self.tp_percentage = tp_percentage
        self.thp_attempts = thp_attempts
        self.thp_percentage = thp_percentage
        self.points_per_game = points_per_game
        self.rebounds_per_game = rebounds_per_game
        self.assists_per_game = assists_per_game
        self.steals_per_game = steals_per_game
        self.blocks_per_game = blocks_per_game
        self.turnovers_per_game = turnovers_per_game
        self._list.append(self)
    
    def __repr__(self):
        return f'<Player: {self.name}>'
    
    def __str__(self):
        return f'({self.team}) {self.name}'
```
